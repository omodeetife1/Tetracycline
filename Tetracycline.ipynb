```python
import pandas as pd
```

```python
# Reading the dataset
tetracycline_data = pd.read_csv('Tetracycline1.csv')
```

```python
# Remove column name 'Unnamed: 0'
tetracycline_data.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
# Create a bar chart of tetracycline concentration over time
plt.bar(tetracycline_data['Time'], tetracycline_data['conc'])

# Setting the X and Y labels
plt.xlabel('Time (hours)')
plt.ylabel('Tetracycline Concentration (mg/L)')

# Adding Title to the bar chart
plt.title('Tetracycline Absorption Profile')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a line plot of tetracycline concentration over time
plt.plot(tetracycline_data['Time'], tetracycline_data['conc'])

# Setting the X and Y labels
plt.xlabel('Time (hours)')
plt.ylabel('Tetracycline Concentration (mg/L)')

# Adding Title to the line plot
plt.title('Tetracycline Elimination Profile')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Bar chart showing conc against Subject
plt.bar(tetracycline_data['conc'], tetracycline_data['Subject'])
 
# Adding Title to the Bar chart
plt.title("Relationship between conc and Subject")
 
# Setting the X and Y labels
plt.xlabel('conc')
plt.ylabel('Subject')
 
plt.show()
```
